# Organize LH sample for PR linTdep LHS multistart

```python
import pandas as pd
import numpy as np

import sys
sys.path.append('../../')
```

```python
PR_no_results = pd.read_csv('Data/Fits/LHS_Fits/PR_noTdep_LHS_fromOpt1_78_.csv',index_col=0)
```

```python
PR_no_results.dropna(axis=0)
```

```python
PR_no_results[PR_no_results.SSR == PR_no_results.SSR.min()]
```

```python
no_46 = PR_no_results.iloc[46]
no_127 = PR_no_results.iloc[127]
```

```python
lhs = pd.read_csv('../../LHS_250x8.txt')
```

```python
lhs['sc_param3'] = ((lhs['Param3']-lhs['Param3'].min())/(lhs['Param3'].max()-lhs['Param3'].min()))*2+(-1)
lhs['sc_param4'] = ((lhs['Param4']-lhs['Param4'].min())/(lhs['Param4'].max()-lhs['Param4'].min()))*2+(-1)
```

```python
no_46_sc = no_46[['sc_param1','sc_param2']]
no_127_sc = no_127[['sc_param1','sc_param2']]
```

```python
no_46_sc_df = pd.DataFrame(no_46_sc).T
sc_46 = pd.concat([no_46_sc_df]*250,ignore_index=True)
sc_46_lhs = pd.concat([sc_46,lhs[['sc_param3','sc_param4']]],axis=1,ignore_index=True)
sc_46_lhs.columns = ['sc_param1','sc_param2','sc_param3','sc_param4']
sc_46_lhs.to_csv('Data/Fits/LHS_Fits/PR_linTdep_46_scLHS.csv',sep=',')
```

```python
no_127_sc_df = pd.DataFrame(no_127_sc).T
sc_127 = pd.concat([no_127_sc_df]*250,ignore_index=True)
sc_127_lhs = pd.concat([sc_127,lhs[['sc_param3','sc_param4']]],axis=1,ignore_index=True)
sc_127_lhs.columns = ['sc_param1','sc_param2','sc_param3','sc_param4']
sc_127_lhs.to_csv('Data/Fits/LHS_Fits/PR_linTdep_127_scLHS.csv',sep=',')
```
