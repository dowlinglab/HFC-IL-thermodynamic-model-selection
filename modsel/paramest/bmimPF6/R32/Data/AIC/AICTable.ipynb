# AIC Table of Models for R32/bmimpf6 system

```python
import pandas as pd
import numpy as np
```

## Import AIC Data

```python
PR_2parameter = pd.read_csv('PR_AIC_2parameter.csv',index_col=0)
PR_2parameter['name'] = 'PR_2parameter'

PR_3parameter_HFCILB0 = pd.read_csv('PR_AIC_3parameter_HFCILB0.csv',index_col=0)
PR_3parameter_HFCILB0['name'] = 'PR_3parameter_HFCILB0'

PR_3parameter_ILHFCB0 = pd.read_csv('PR_AIC_3parameter_ILHFCB0.csv',index_col=0)
PR_3parameter_ILHFCB0['name'] = 'PR_3parameter_ILHFCB0'

PR_4parameter = pd.read_csv('PR_AIC_4parameter.csv',index_col=0)
PR_4parameter['name'] = 'PR_4parameter'

PR_6parameter = pd.read_csv('PR_AIC_6parameter.csv',index_col=0)
PR_6parameter['name'] = 'PR_6parameter'
```

```python
SRK_1parameter_ILHFCfit = pd.read_csv('SRK_AIC_1parameter_ILHFCfit.csv',index_col=0)
SRK_1parameter_ILHFCfit['name'] = 'SRK_1parameter_ILHFCfit'

SRK_2parameter = pd.read_csv('SRK_AIC_2parameter.csv',index_col=0)
SRK_2parameter['name'] = 'SRK_2parameter'

SRK_3parameter_HFCILB0 = pd.read_csv('SRK_AIC_3parameter_HFCILB0.csv',index_col=0)
SRK_3parameter_HFCILB0['name'] = 'SRK_3parameter_HFCILB0'

SRK_3parameter_ILHFCB0 = pd.read_csv('SRK_AIC_3parameter_ILHFCB0.csv',index_col=0)
SRK_3parameter_ILHFCB0['name'] = 'SRK_3parameter_ILHFCB0'

SRK_4parameter = pd.read_csv('SRK_AIC_4parameter.csv',index_col=0)
SRK_4parameter['name'] = 'SRK_4parameter'

SRK_6parameter = pd.read_csv('SRK_AIC_6parameter.csv',index_col=0)
SRK_6parameter['name'] = 'SRK_6parameter'
```

```python
PR = pd.concat([PR_2parameter,PR_3parameter_HFCILB0,PR_3parameter_ILHFCB0,PR_4parameter,PR_6parameter])
```

```python
SRK = pd.concat([SRK_1parameter_ILHFCfit,SRK_2parameter,SRK_3parameter_HFCILB0,SRK_3parameter_ILHFCB0,SRK_4parameter,SRK_6parameter])
```

```python
AICTable = pd.concat([PR,SRK],ignore_index=True)
```

```python
AICTable
```

```python
AICTable['Relative_Likelihood'] = np.exp((AICTable['AIC'].min() - AICTable['AIC'])/2)
```

```python
AICTable
```

```python
AICTable_Ranked = AICTable.sort_values(['Relative_Likelihood'],ascending=False,ignore_index=True)
```
