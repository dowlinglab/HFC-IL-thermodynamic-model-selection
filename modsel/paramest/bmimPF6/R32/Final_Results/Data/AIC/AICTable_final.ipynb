# AIC Table of Models for R32/bmimPF6 system

```python
import pandas as pd
import numpy as np
```

## Import AIC Data

```python
PR_1param_Opt1 = pd.read_csv('PR_AIC_1param_Opt1.csv',index_col=0)
PR_1param_Opt1['name'] = 'PR_1param_Opt1'

PR_1param_Opt2 = pd.read_csv('PR_AIC_1param_Opt2.csv',index_col=0)
PR_1param_Opt2['name'] = 'PR_1param_Opt2'

PR_noTdep = pd.read_csv('PR_AIC_noTdep.csv',index_col=0)
PR_noTdep['name'] = 'PR_noTdep'

PR_3params_Opt1 = pd.read_csv('PR_AIC_3params_Opt1.csv',index_col=0)
PR_3params_Opt1['name'] = 'PR_3params_Opt1'

PR_3params_Opt2 = pd.read_csv('PR_AIC_3params_Opt2.csv',index_col=0)
PR_3params_Opt2['name'] = 'PR_3params_Opt2'

PR_linTdep = pd.read_csv('PR_AIC_linTdep.csv',index_col=0)
PR_linTdep['name'] = 'PR_linTdep'

PR_quadTdep = pd.read_csv('PR_AIC_quadTdep.csv',index_col=0)
PR_quadTdep['name'] = 'PR_quadTdep'

PR_polyTdep = pd.read_csv('PR_AIC_polyTdep.csv',index_col=0)
PR_polyTdep['name'] = 'PR_polyTdep'
```

```python
SRK_1param_Opt1 = pd.read_csv('SRK_AIC_1param_Opt1.csv',index_col=0)
SRK_1param_Opt1['name'] = 'SRK_1param_Opt1'

SRK_1param_Opt2 = pd.read_csv('SRK_AIC_1param_Opt2.csv',index_col=0)
SRK_1param_Opt2['name'] = 'SRK_1param_Opt2'

SRK_noTdep = pd.read_csv('SRK_AIC_noTdep.csv',index_col=0)
SRK_noTdep['name'] = 'SRK_noTdep'

SRK_3params_Opt1 = pd.read_csv('SRK_AIC_3params_Opt1.csv',index_col=0)
SRK_3params_Opt1['name'] = 'SRK_3params_Opt1'

SRK_3params_Opt2 = pd.read_csv('SRK_AIC_3params_Opt2.csv',index_col=0)
SRK_3params_Opt2['name'] = 'SRK_3params_Opt2'

SRK_linTdep = pd.read_csv('SRK_AIC_linTdep.csv',index_col=0)
SRK_linTdep['name'] = 'SRK_linTdep'

SRK_quadTdep = pd.read_csv('SRK_AIC_quadTdep.csv',index_col=0)
SRK_quadTdep['name'] = 'SRK_quadTdep'

SRK_polyTdep = pd.read_csv('SRK_AIC_polyTdep.csv',index_col=0)
SRK_polyTdep['name'] = 'SRK_polyTdep'
```

```python
PR = pd.concat([PR_1param_Opt1,PR_1param_Opt2,PR_noTdep,PR_3params_Opt1,PR_3params_Opt2,PR_linTdep,PR_quadTdep,PR_polyTdep])
```

```python
SRK = pd.concat([SRK_1param_Opt1,SRK_1param_Opt2,SRK_noTdep,SRK_3params_Opt1,SRK_3params_Opt2,SRK_linTdep,SRK_quadTdep,SRK_polyTdep])
```

```python
AICTable = pd.concat([PR,SRK],ignore_index=True)
```

```python
AICTable
```

```python
AICTable['Relative_Log_Likelihood'] = (AICTable['AIC'].min() - AICTable['AIC'])/2
```

```python
AICTable
```

```python
AICTable_Ranked = AICTable.sort_values(['Relative_Log_Likelihood'],ascending=False,ignore_index=True)
```
