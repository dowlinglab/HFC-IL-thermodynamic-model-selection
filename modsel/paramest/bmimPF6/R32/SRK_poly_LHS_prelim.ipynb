# Organize LH sample for PR poly LHS multistart

```python
import pandas as pd
import numpy as np

import sys
sys.path.append('../../')
```

```python
SRK_quad_results = pd.read_csv('Final_Results/Data/Init_Final/LHS_Output/SRK_quadTdep_LHS_final.csv',index_col=0)
```

```python
SRK_quad_results.dropna(axis=0)
```

```python
quad = SRK_quad_results.iloc[1501]
```

```python
lhs = pd.read_csv('../../LHS_5000x8.txt')
```

```python
lhs['sc_param7'] = ((lhs['Param7']-lhs['Param7'].min())/(lhs['Param7'].max()-lhs['Param7'].min()))*4+(-2)
lhs['sc_param8'] = ((lhs['Param8']-lhs['Param8'].min())/(lhs['Param8'].max()-lhs['Param8'].min()))*4+(-2)
```

```python
quad_sc = quad[['sc_param1','sc_param2','sc_param3','sc_param4','sc_param5','sc_param6']]
# quad_227_act = quad_227[['PR_kappa_A[emimTf2N,R32]','PR_kappa_A[R32,emimTf2N]','PR_kappa_B[emimTf2N,R32]','PR_kappa_B[R32,emimTf2N]','PR_kappa_C[emimTf2N,R32]','PR_kappa_C[R32,emimTf2N]']]
```

```python
quad_sc_df = pd.DataFrame(quad_sc).T
sc = pd.concat([quad_sc_df]*5000,ignore_index=True)
sc_lhs = pd.concat([sc,lhs[['sc_param7','sc_param8']]],axis=1,ignore_index=True)
sc_lhs.columns = ['sc_param1','sc_param2','sc_param3','sc_param4','sc_param5','sc_param6','sc_param7','sc_param8']
sc_lhs.to_csv('Final_Results/Data/Init_Final/SRK_polyTdep_LHS_fromquad.csv',sep=',')
```
