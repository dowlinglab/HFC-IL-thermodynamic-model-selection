# Optimality Table of Models for R32/emimTF2N system

```python
import pandas as pd
import numpy as np
```

## Import Optimality Data

```python
PR_1param_Opt1 = pd.read_csv('PR_FIMtrdet_1param_Opt1.csv',header=None).T
PR_1param_Opt1.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_1param_Opt1['name'] = 'PR_1param_Opt1'

PR_1param_Opt2 = pd.read_csv('PR_FIMtrdet_1param_Opt2.csv',header=None).T
PR_1param_Opt2.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_1param_Opt2['name'] = 'PR_1param_Opt2'

PR_1param_Opt2_GF = pd.read_csv('PR_FIMtrdet_1param_Opt2_GF.csv',header=None).T
PR_1param_Opt2_GF.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_1param_Opt2_GF['name'] = 'PR_1param_Opt2_GF'

PR_noTdep = pd.read_csv('PR_FIMtrdet_noTdep.csv',header=None).T
PR_noTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_noTdep['name'] = 'PR_noTdep'

PR_3params_Opt1 = pd.read_csv('PR_FIMtrdet_3params_Opt1.csv',header=None).T
PR_3params_Opt1.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_3params_Opt1['name'] = 'PR_3params_Opt1'

PR_3params_Opt2 = pd.read_csv('PR_FIMtrdet_3params_Opt2.csv',header=None).T
PR_3params_Opt2.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_3params_Opt2['name'] = 'PR_3params_Opt2'

PR_linTdep = pd.read_csv('PR_FIMtrdet_linTdep.csv',header=None).T
PR_linTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_linTdep['name'] = 'PR_linTdep'

PR_quadTdep = pd.read_csv('PR_FIMtrdet_quadTdep.csv',header=None).T
PR_quadTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_quadTdep['name'] = 'PR_quadTdep'

PR_polyTdep = pd.read_csv('PR_FIMtrdet_polyTdep.csv',header=None).T
PR_polyTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
PR_polyTdep['name'] = 'PR_polyTdep'
```

```python
SRK_1param_Opt1 = pd.read_csv('SRK_FIMtrdet_1param_Opt1.csv',header=None).T
SRK_1param_Opt1.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
SRK_1param_Opt1['name'] = 'SRK_1param_Opt1'

SRK_1param_Opt2 = pd.read_csv('SRK_FIMtrdet_1param_Opt2.csv',header=None).T
SRK_1param_Opt2.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval'] 
SRK_1param_Opt2['name'] = 'SRK_1param_Opt2'

SRK_noTdep = pd.read_csv('SRK_FIMtrdet_noTdep.csv',header=None).T
SRK_noTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
SRK_noTdep['name'] = 'SRK_noTdep'

SRK_3params_Opt1 = pd.read_csv('SRK_FIMtrdet_3params_Opt1.csv',header=None).T
SRK_3params_Opt1.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
SRK_3params_Opt1['name'] = 'SRK_3params_Opt1'

SRK_3params_Opt2 = pd.read_csv('SRK_FIMtrdet_3params_Opt2.csv',header=None).T
SRK_3params_Opt2.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval'] 
SRK_3params_Opt2['name'] = 'SRK_3params_Opt2'

SRK_linTdep = pd.read_csv('SRK_FIMtrdet_linTdep.csv',header=None).T
SRK_linTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
SRK_linTdep['name'] = 'SRK_linTdep'

SRK_quadTdep = pd.read_csv('SRK_FIMtrdet_quadTdep.csv',header=None).T
SRK_quadTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
SRK_quadTdep['name'] = 'SRK_quadTdep'

SRK_polyTdep = pd.read_csv('SRK_FIMtrdet_polyTdep.csv',header=None).T
SRK_polyTdep.columns = ['cov_trace','FIM_trace','determinant','min_eigval','max_eigval']
SRK_polyTdep['name'] = 'SRK_polyTdep'
```

```python
PR = pd.concat([PR_1param_Opt1,PR_1param_Opt2,PR_1param_Opt2_GF,PR_noTdep,PR_3params_Opt1,PR_3params_Opt2,PR_linTdep,PR_quadTdep,PR_polyTdep])
```

```python
SRK = pd.concat([SRK_1param_Opt1,SRK_1param_Opt2,SRK_noTdep,SRK_3params_Opt1,SRK_3params_Opt2,SRK_linTdep,SRK_quadTdep,SRK_polyTdep])
```

```python
OptTable = pd.concat([PR,SRK],ignore_index=True)
```

```python
OptTable['CN'] = OptTable['max_eigval']/OptTable['min_eigval']
```

```python
OptTable
```

```python
OptTable_Aopt = OptTable.sort_values(['cov_trace'],ascending=True,ignore_index=True)
```

```python
OptTable_Dopt = OptTable.sort_values(['determinant'],ascending=False,ignore_index=True)
```

```python
OptTable_Eopt = OptTable.sort_values(['min_eigval'],ascending=False,ignore_index=True)
```
