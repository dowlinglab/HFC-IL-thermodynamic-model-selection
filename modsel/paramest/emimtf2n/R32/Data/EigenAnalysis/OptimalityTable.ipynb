# Optimality Table of Models for R32/emimTF2N system

```python
import pandas as pd
import numpy as np
```

## Import Optimality Data

```python
PR_1parameter_HFCILfit = pd.read_csv('PR_FIMtrdet_1parameter_HFCILfit.csv',header=None).T
PR_1parameter_HFCILfit.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_1parameter_HFCILfit['name'] = 'PR_1parameter_HFCILfit'

PR_1parameter_ILHFCfit = pd.read_csv('PR_FIMtrdet_1parameter_ILHFCfit.csv',header=None).T
PR_1parameter_ILHFCfit.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_1parameter_ILHFCfit['name'] = 'PR_1parameter_ILHFCfit'

PR_1parameter_ILHFCfit_GF = pd.read_csv('PR_FIMtrdet_1parameter_ILHFCfit_GF.csv',header=None).T
PR_1parameter_ILHFCfit_GF.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_1parameter_ILHFCfit_GF['name'] = 'PR_1parameter_ILHFCfit_GF'

PR_2parameter = pd.read_csv('PR_FIMtrdet_2parameter.csv',header=None).T
PR_2parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_2parameter['name'] = 'PR_2parameter'

PR_3parameter_HFCILB0 = pd.read_csv('PR_FIMtrdet_3parameter_HFCILB0.csv',header=None).T
PR_3parameter_HFCILB0.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_3parameter_HFCILB0['name'] = 'PR_3parameter_HFCILB0'

PR_3parameter_ILHFCB0 = pd.read_csv('PR_FIMtrdet_3parameter_ILHFCB0.csv',header=None).T
PR_3parameter_ILHFCB0.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_3parameter_ILHFCB0['name'] = 'PR_3parameter_ILHFCB0'

PR_4parameter = pd.read_csv('PR_FIMtrdet_4parameter.csv',header=None).T
PR_4parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_4parameter['name'] = 'PR_4parameter'

PR_6parameter = pd.read_csv('PR_FIMtrdet_6parameter.csv',header=None).T
PR_6parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_6parameter['name'] = 'PR_6parameter'

PR_8parameter = pd.read_csv('PR_FIMtrdet_8parameter.csv',header=None).T
PR_8parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
PR_8parameter['name'] = 'PR_8parameter'
```

```python
SRK_1parameter_HFCILfit = pd.read_csv('SRK_FIMtrdet_1parameter_HFCILfit.csv',header=None).T
SRK_1parameter_HFCILfit.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
SRK_1parameter_HFCILfit['name'] = 'SRK_1parameter_HFCILfit'

SRK_1parameter_ILHFCfit = pd.read_csv('SRK_FIMtrdet_1parameter_ILHFCfit.csv',header=None).T
SRK_1parameter_ILHFCfit.columns = ['cov_trace','FIM_trace','determinant','min_eigval'] 
SRK_1parameter_ILHFCfit['name'] = 'SRK_1parameter_ILHFCfit'

SRK_2parameter = pd.read_csv('SRK_FIMtrdet_2parameter.csv',header=None).T
SRK_2parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
SRK_2parameter['name'] = 'SRK_2parameter'

SRK_3parameter_HFCILB0 = pd.read_csv('SRK_FIMtrdet_3parameter_HFCILB0.csv',header=None).T
SRK_3parameter_HFCILB0.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
SRK_3parameter_HFCILB0['name'] = 'SRK_3parameter_HFCILB0'

SRK_3parameter_ILHFCB0 = pd.read_csv('SRK_FIMtrdet_3parameter_ILHFCB0.csv',header=None).T
SRK_3parameter_ILHFCB0.columns = ['cov_trace','FIM_trace','determinant','min_eigval'] 
SRK_3parameter_ILHFCB0['name'] = 'SRK_3parameter_ILHFCB0'

SRK_4parameter = pd.read_csv('SRK_FIMtrdet_4parameter.csv',header=None).T
SRK_4parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
SRK_4parameter['name'] = 'SRK_4parameter'

SRK_6parameter = pd.read_csv('SRK_FIMtrdet_6parameter.csv',header=None).T
SRK_6parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
SRK_6parameter['name'] = 'SRK_6parameter'

SRK_8parameter = pd.read_csv('SRK_FIMtrdet_8parameter.csv',header=None).T
SRK_8parameter.columns = ['cov_trace','FIM_trace','determinant','min_eigval']
SRK_8parameter['name'] = 'SRK_8parameter'
```

```python
PR = pd.concat([PR_1parameter_HFCILfit,PR_1parameter_ILHFCfit,PR_1parameter_ILHFCfit_GF,PR_2parameter,PR_3parameter_HFCILB0,PR_3parameter_ILHFCB0,PR_4parameter,PR_6parameter,PR_8parameter])
```

```python
SRK = pd.concat([SRK_1parameter_HFCILfit,SRK_1parameter_ILHFCfit,SRK_2parameter,SRK_3parameter_HFCILB0,SRK_3parameter_ILHFCB0,SRK_4parameter,SRK_6parameter,SRK_8parameter])
```

```python
OptTable = pd.concat([PR,SRK],ignore_index=True)
```

```python
OptTable
```

```python
OptTable_Aopt = OptTable.sort_values(['cov_trace'],ascending=True,ignore_index=True)
```

```python
OptTable_Dopt = OptTable.sort_values(['determinant'],ascending=False,ignore_index=True)
```

```python
OptTable_Eopt = OptTable.sort_values(['min_eigval'],ascending=False,ignore_index=True)
```
