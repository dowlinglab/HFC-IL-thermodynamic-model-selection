# Plotting GF's BIP Temperature Dependence Findings

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## kij
K12 = [-0.02373446, -0.03992035, -0.06142196, -0.0686277]
K21 = [-0.04897066, -0.17013375, -0.31491854, -0.35055943]
T = [283.1, 298.1, 323.1, 348.1]

fig, ax = plt.subplots()

ax.plot(T,K12,marker='*',linestyle='',color='blue',label="k$_{ij}$") 
ax.plot(T,K21,marker='x',linestyle='',color='red',label="k$_{ji}$") 

ax.set_ylabel('Parameter Values',fontsize=24)
ax.set_xlabel('Temperature, K',fontsize=24)

ax.set_xticks([280, 300, 320, 340, 360])
ax.set_xticks([290, 310, 330, 350],minor=True)
ax.set_yticks([-0.4, -0.3, -0.2, -0.1, 0.0])
ax.set_yticks([-0.35, -0.25, -0.15, -0.05],minor=True)

ax.tick_params("both", direction="in", which="both", length=4, labelsize=20, pad=10)
ax.tick_params("both", which="major", length=8)
ax.xaxis.set_ticks_position("both")
ax.yaxis.set_ticks_position("both")

ax.legend(fontsize=24, handletextpad=0.00, loc="lower left")#, bbox_to_anchor=(1, 0.5))
fig.savefig('R32-emimTF2N-Tdepex.pdf',bbox_inches='tight')
plt.show()
```
