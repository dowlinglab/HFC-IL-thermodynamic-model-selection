```python
import numpy as np
import csv
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.optimize import leastsq

import scipy.stats as stats


from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms
import matplotlib.patches as mpatches
```

```python
def make_ellipse(mu, cov, ci=0.95):
    e, v = np.linalg.eig(cov)
  
    alpha = np.arctan(v[0,1]/v[0,0])
    angle = -alpha*360/np.pi 
    q = stats.chi2(2).ppf(ci)
    e = Ellipse(mu, 2*np.sqrt(q*e[0]), 2*np.sqrt(q*e[1]), angle=angle)
    return e

cov2 = np.array([[0.000159, 0.000017],
                [0.000017, 0.000002]])

mu = -0.101928, -0.032234

e = make_ellipse(mu, cov2, ci=0.95)

e.set_alpha(0.5)
e.set_facecolor('red')
e.set_zorder(10)

ax = plt.subplot(111)

ax.add_artist(e)

plt.title('95% confidence ellpise')
plt.xlabel("$\kappa_{12}$")
plt.ylabel("$\kappa_{21}$")

red_patch = mpatches.Patch(color='red', label='Frequentist approach',alpha=1)

plt.legend(handles=[red_patch])
# plt.savefig('ellipses.png')
ax.set(xlim=[-0.05, -0.15], ylim=[-0.025, -0.04])
plt.show()
```
