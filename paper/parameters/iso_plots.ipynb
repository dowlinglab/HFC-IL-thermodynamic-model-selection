```python
# Import plotting functions
import matplotlib.pyplot as plt

# Import numpy library 
import numpy as np

import pandas as pd
```

```python
data = pd.read_csv('R125IL.csv')
data1 = pd.read_csv('P_lin_125.csv')
data2 = pd.read_csv('P_cons_125.csv')
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(1, 1, 1)

plt.plot(data["x_R125"][0:8],data1["0"][0:8]/1e6,"b-",label="kappa T dependent T= 283 K")
plt.plot(data["x_R125"][8:16],data1["0"][8:16]/1e6,"b--",label="kappa T dependent T= 298 K")  
plt.plot(data["x_R125"][16:24],data1["0"][16:24]/1e6,"b-.",label="kappa T dependent T= 323 K")  
plt.plot(data["x_R125"][24:],data1["0"][24:]/1e6,"b:",label="kappa T dependent T= 348 K")  

plt.plot(data["x_R125"][0:8],data2["0"][0:8]/1e6,"r-",label="kappa constant T= 283 K")
plt.plot(data["x_R125"][8:16],data2["0"][8:16]/1e6,"r--",label="kappa constant T= 298 K")  
plt.plot(data["x_R125"][16:24],data2["0"][16:24]/1e6,"r-.",label="kappa constant T= 323 K")  
plt.plot(data["x_R125"][24:],data2["0"][24:]/1e6,"r:",label="kappa constant T= 348 K")  

plt.plot(data["x_R125"],data["pressure"]/1e6,"*",color='orange',label="data")  
plt.title('Isotherm R125-[bmim][PF6]',fontsize=20)
plt.ylabel('Pressure (MPa)',fontsize=20)
plt.xlabel('x R125',fontsize=20)
plt.grid(True)
plt.ylim(0.1,1.110000)
plt.legend()
plt.savefig('iso125.png')

plt.show()
```
