```python
import pandas as pd

# Import plotting functions
import matplotlib.pyplot as plt

# Import numpy library 
import numpy as np
```

```python
df = pd.read_csv('ku_100.csv')
```

```python
IL_xe = [0.910978,0.920514854,0.938673956,0.955980518,0.972832392]
R32_xe = [0.3405,0.2576,0.1687,0.0835,0.0001]
R125_xe = [0,0.0221,0.0471,0.0714,0.0955]
IL_ye = [0, 0, 0, 0, 0]
R32_ye = [1,0.75,0.5,0.25,0]
R125_ye = [0,0.25,0.5,0.75,1]

Rtot = [sum(x) for x in zip(R32_xe, R125_xe )]

a = df['R125_x']*1000/(284*(1-df['R125_x'])+120*df['R125_x'])
a2 = df['R125_x2']*1000/(284*(1-df['R125_x2'])+120*df['R125_x2'])
a3 = df['R125_x3']*1000/(284*(1-df['R125_x3'])+120*df['R125_x3'])

b = df['R32_x']*1000/(284*(1-df['R32_x'])+52*df['R32_x'])
b2 = df['R32_x2']*1000/(284*(1-df['R32_x2'])+52*df['R32_x2'])
b3 = df['R32_x3']*1000/(284*(1-df['R32_x3'])+52*df['R32_x3'])

c = a+b
c2 = a2+b2
c3 = a3+b3



fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1)

plt.plot(df['R125_y'],a,"bx",label="HFC-125 Ternary") 
plt.plot(df['R125_y2'],a2,"b*",label="HFC-125 Binary") 
plt.plot(df['R125_y3'],a3,"bo",label="HFC-125 Binary T dep") 
plt.plot(R125_ye,R125_xe,"b-",label="HFC-125 experimental")  

plt.plot(df['R125_y'],b,"rx",label="HFC-32 Ternary") 
plt.plot(df['R125_y2'],b2,"r*",label="HFC-32 Binary") 
plt.plot(df['R125_y3'],b3,"ro",label="HFC-32 Binary T dep") 
plt.plot(R125_ye,R32_xe,"r-",label="HFC-32 experimental") 

plt.plot(df['R125_y'],c,"gx",label="TotalTernary") 
plt.plot(df['R125_y2'],c2,"g*",label="Total Binary") 
plt.plot(df['R125_y3'],c3,"go",label="Total Binary T dep") 
plt.plot(R125_ye,Rtot,"g:",label="Total experimental") 
plt.title('Adsorption at 100kPa, 25C',fontsize=20)
plt.ylabel('Adsorption (mmol/g)',fontsize=20)
plt.xlabel('HFC-125 Inlet Mole Fraction',fontsize=20)
plt.grid(True)
# plt.ylim(0,1.110000)
# plt.xlim(0,0.5)
plt.legend()
plt.savefig('ads_250.png')
plt.show()
```
