```python
matplotlib.rcParams['figure.dpi'] = 200
matplotlib.rcParams['figure.figsize'] = (8, 8)    
    
scale = 100
figure, tax = ternary.figure(scale=scale)

# Draw Boundary and Gridlines
tax.boundary(linewidth=2.0)
tax.gridlines(color="blue", multiple=5)

fontsize = 20
offset = 0.10
# tax.set_title("Ternary phase diagram 8 bar\n", fontsize=24)
tax.left_axis_label("HFC-32 [mole %]", fontsize=fontsize, offset=offset)
tax.right_axis_label("HFC-125 [mole %]", fontsize=fontsize, offset=offset)
tax.bottom_axis_label("[bmim][PF6] [mole %]", fontsize=fontsize, offset=0)

tax.line(points_x1[0], points_x1[4], linewidth=1, color='red', linestyle=linestyle[0], label="M$_B$")

tax.line(points_x2[0], points_x2[4], linewidth=1, color='blue', linestyle=linestyle[1], label="M$_{B,K}$")

tax.line(points_x3[0], points_x3[4], linewidth=1, color='green', linestyle=linestyle[2], label="M$_T$")

tax.line(points_x4[0], points_x4[4], linewidth=1, color='purple', linestyle=linestyle[3],label="M$_{BT}$")

tax.line(points_x5[0], points_x5[4], linewidth=1, color='black', linestyle=linestyle[4],label="M$_{BT,K}$")

tax.scatter(points_x, color='cyan', marker="o", s=40, label="Ternary data")

tax.legend(fontsize=14)
tax.ticks(axis='lbr', multiple=10, linewidth=1, fontsize=14, offset=0.015)
tax.get_axes().axis('off')
tax.clear_matplotlib_ticks()
tax.savefig('all_ter.jpeg', bbox_inches='tight',dpi=300)
tax.show()
```
