```python
import pandas as pd

# Import plotting functions
import matplotlib.pyplot as plt

# Import numpy library 
import numpy as np

import ternary
import matplotlib
```

```python
data1 = pd.read_csv('ternary_ternary2.csv')
data2 = pd.read_csv('ternary_linear.csv')
data3 = pd.read_csv('ternary_const.csv')
```

```python
IL_xd = [0.794729917,0.822135707,0.851896931,0.881890878,0.919985062]
R32_xd = [0.205270083,0.162225144,0.11537989,0.067126175,8.81E-05]
R125_xd = [0,0.015639149,0.032723178,0.050982947,0.080103028]
```

```python
points_x = []

for i in range(len(IL_xd)):
    points = (IL_xd[i]*100, R32_xd[i]*100, R125_xd[i]*100)
    points_x.append(points)

# points_y = []

# for i in range(len(IL_xd)):
#     points = (data1["IL_y"][i]*100, data1["R32_y"][i]*100, data1["R125_y"][i]*100)
#     points_y.append(points)
```

```python
points_x1 = []

for i in range(len(data1["IL_x"])):
    points = (data1["IL_x"][i]*100, data1["R32_x"][i]*100, data1["R125_x"][i]*100)
    points_x1.append(points)

points_y1 = []

for i in range(len(data1["IL_x"])):
    points = (data1["IL_y"][i]*100, data1["R32_y"][i]*100, data1["R125_y"][i]*100)
    points_y1.append(points)
```

```python
points_x2 = []

for i in range(len(data2["IL_x"])):
    points = (data2["IL_x"][i]*100, data2["R32_x"][i]*100, data2["R125_x"][i]*100)
    points_x2.append(points)

points_y2 = []

for i in range(len(data2["IL_x"])):
    points = (data2["IL_y"][i]*100, data2["R32_y"][i]*100, data2["R125_y"][i]*100)
    points_y2.append(points)
```

```python
points_x3 = []

for i in range(len(data3["IL_x"])):
    points = (data3["IL_x"][i]*100, data3["R32_x"][i]*100, data3["R125_x"][i]*100)
    points_x3.append(points)

points_y3 = []

for i in range(len(data3["IL_x"])):
    points = (data3["IL_y"][i]*100, data3["R32_y"][i]*100, data3["R125_y"][i]*100)
    points_y3.append(points)
```

```python
matplotlib.rcParams['figure.dpi'] = 200
matplotlib.rcParams['figure.figsize'] = (8, 8)    
    
scale = 100
figure, tax = ternary.figure(scale=scale)

# Draw Boundary and Gridlines
tax.boundary(linewidth=2.0)
tax.gridlines(color="blue", multiple=5)

fontsize = 12
offset = 0.14
# tax.set_title("Ternary phase diagram 8 bar\n", fontsize=24)
tax.left_axis_label("R32 mole percent", fontsize=fontsize, offset=offset)
tax.right_axis_label("R125 mole percent", fontsize=fontsize, offset=offset)
tax.bottom_axis_label("[bmim][PF6] mole percent", fontsize=fontsize, offset=offset)

#for pressure changing plot would need to draw line between x[0] and x[final] for the indivual pressure
#use tax.line with the tax.scatter below

tax.line(points_x1[0], points_x1[4], linewidth=1., color='green', linestyle="-")

tax.line(points_x2[0], points_x2[4], linewidth=1., color='blue', linestyle="-")

tax.line(points_x3[0], points_x3[4], linewidth=1., color='red', linestyle="-")

# tax.line(points_x[0], points_x[4], linewidth=1., color='red', linestyle=":")

# make this tax.line
tax.scatter(points_x, marker='s', color='orange', label="liquid composition")
# tax.scatter(points_x1, marker='d', color='blue', label="liquid composition")
# #Comment out for pressure changing plot
# tax.scatter(points_y1, marker='s', color='blue', label="vapor composition")
# tax.legend()
tax.ticks(axis='lbr', multiple=10, linewidth=1, offset=0.025)
tax.get_axes().axis('off')
tax.clear_matplotlib_ticks()
tax.savefig('all_ter.pdf')
tax.show()
```
